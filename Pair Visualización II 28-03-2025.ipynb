
Ejercicio 1: ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.


```python

# Contar cuántos países hay en cada continente
continent_counts = df['continent'].value_counts()

# Obtener el continente con más países
top_continent = continent_counts.idxmax()
top_continent_count = continent_counts.max()

# Calcular el porcentaje
total_countries = continent_counts.sum()
percentage = (top_continent_count / total_countries) * 100

# Mostrar resultados
print(f"El continente con más países es {top_continent}, con {top_continent_count} países.")
print(f"Representa el {percentage:.2f}% del total de países en el conjunto de datos.")
```


Ejercicio 2: ¿Cuáles son los 10 países con la mayor área forestal?

```python

# Suponiendo que la columna del área forestal se llama "area_forestal"
# Si el nombre es diferente, cámbialo por el correcto
df_sorted = df[['country', 'forested_area']].dropna().sort_values(by='forested_area', ascending=False)

# Obtener los 10 países con mayor área forestal
top_10_forest = df_sorted.head(10)

# Mostrar resultados
print("Los 10 países con mayor área forestal:")
print(top_10_forest)
```

```python
# GRAFICA:
plt.figure(figsize=(6,6))
sns.barplot(top_10_forest, x='forested_area', y='country', palette='viridis')
plt.xlabel("Área forestal (%)")
plt.ylabel("País")
plt.title("Top 10 países con mayor área forestal");
```
