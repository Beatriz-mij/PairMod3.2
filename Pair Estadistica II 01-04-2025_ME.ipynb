```python
for continent in df["life_expectancy"].unique():
    df_life = df[df["life_expectancy"] == continent]
    media = df_life["life_expectancy"].mean()
    error = stats.sem(df_life["life_expectancy"])
    grados_libertad = len(df_life["life_expectancy"])-1
    nivel_confianza = 0.95
    valor_critico = stats.t.ppf((1 + nivel_confianza) / 2, df=grados_libertad)
    salto = valor_critico * error
    limite_sup = media + salto
    limite_inf = media - salto
    print(continent)
    print(limite_inf, "-", limite_sup)
    print("---------------------------------------------------------")
```
